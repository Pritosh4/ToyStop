## So which toy was the best seller?
Since we don't have the actual data of how many of each toy has been sold, we'll use the number of reviews that each toys has since every reviewer who has given a review would have bough the toy.

```python
%%sql
SELECT *
FROM amazon_data
ORDER BY reviews DESC
LIMIT 1;
```

As we can see above, the best seller isn't really a very sophiticated toy but really just a simple toy to stack rings!

## Find the best rated toys
On the Amazon India if you sort the products on the basis of their average rating, it will not give you an appropriate answer because average rating works fine if you always have a lot of ratings, but suppose item A has 2 five star rating. Suppose item B has 100 five star ratings and 1 one star rating. This algorithm puts item B (lots of five star ratings) below item A (very few five star ratings). 

Hence instead what we'll do is that we'll sort the data we have on the basis of the ratings, provided that the toy has more than a pre-determined number (500 here) of reviews.

```python
%%sql 
SELECT *
FROM amazon_data
WHERE rating >= 4.5 AND reviews >= 500
ORDER BY rating DESC, price;
```

As we can see there are actually only 7 toys that have managed to keep a rating above 4.5 even after selling more than 500 units.

## Manufacturer recommended age
If you would have noticed, we have a column that mentions the Manufacturer recommended age for a particular toy. This is basically the minimum age suggested by the manufacturer that the child has to be in order to play with the toy.

Let's take a look at the different Manufacturer recommended ages we have in our dataset.

```python
%%sql 
SELECT DISTINCT("Manufacturer recommended age")
FROM amazon_data
ORDER BY "Manufacturer recommended age";
```

As we can see we have multiple values for the ages, but by just knowing this, it does not really give us an insight into the number of toys that are there for each of the ages or even if some age groups have more toys available as comapred to others.

Hence to tackle this issue let us create age wise bins and calculate how many toys are avilable in our dataset that belong to their particular bins.

```python
%%sql
WITH bins AS (SELECT generate_series(0, 15, 1) AS lower,
              generate_series(1, 16, 1) AS upper),
    items AS (SELECT title, "Manufacturer recommended age"
              FROM amazon_data)

SELECT lower, upper, COUNT(title)
FROM bins
LEFT JOIN items
ON "Manufacturer recommended age" >= lower AND "Manufacturer recommended age" < upper
GROUP BY lower, upper
ORDER BY count DESC;
```

As we can see clearly that the most number of toys are in the 3 - 4 years bin and more than 85% of the toys data that we have, have a Manufacturer recommended age between 0 - 4. 

**So what does this mean?**
While this does not mean that kids older than 4 years (say 10 or 11 year olds) cannot use these toys. It means that if they are given these toys then they might find it too childish for their age and would not be interested to play with them.
Hence customers would find it difficult to have enough options to buy toys for their pre-teenager and will have to look elsewhere.
